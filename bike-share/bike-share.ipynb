```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# According to the link hour.csv and day.csv have the same field (except 'hr' which is not in day.csv)
# hence we will use hour.csv
dataset = pd.read_csv('hour.csv')
print(dataset.head())
print(dataset.shape)
```

```python
dataset.describe()
```

```python
# according to the UCI website, this dataset does not have any missing values
# to confirm we can:
dataset.isnull().values.any()
```

```python
dataset.dtypes
# column dteday has values such as 2011-01-01.
```

```python
# here (from above) we are converting some int data types to categorical
# this is because they are originally categorical (dataset information) 
# but were parsed as int. 

dataset['season'] = dataset.season.astype('category')
dataset['mnth'] = dataset.mnth.astype('category')
dataset['hr'] = dataset.hr.astype('category')
dataset['holiday'] = dataset.holiday.astype('category')
dataset['weekday'] = dataset.weekday.astype('category')
dataset['workingday'] = dataset.workingday.astype('category')
dataset['weathersit'] = dataset.weathersit.astype('category')
```

```python
dataset.hist(figsize=(8,8))
# from the visualisation below - some things do stick out. 
# some data (cnt, causal, registers, windspeed) is left skewed
# whereas, (temp, hum, atemp) is normally distributed. 
# other columns are doing their own thing.
# this could possible show that some regression based models, such as linear regression, may not work.
# we can still experiment with said model but shouldnt expect the best of results. 
```

```python
ax, fig = plt.subplots(figsize=(16,10))
correlation_matrix = dataset.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="OrRd")
plt.show()

# we can see some correlation in the top left and bottom right corner, but after looking at the attribute descriptions, there maybe no link.
# some attributes here are very close to 0 (i.e. hum, atemp, windspeed) as compared to cnt. 
# the attribut 'instant' is just a record index of each data point. We, as humans, know that doesnt hold any value to the 
# dataset and therefore should be removed (will be removed later on)
# but we can see that in some cases it has high correlation (and low) which couild possibly skew the results if kept in. 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.svm import LinearSVC
from sklearn.linear_model import LinearRegression

dataset.shape # to remind us of the original shape
```

since the dataset has 17 attributes some of them may not be as useful to the training process, 
although this conformation can only come with domain expertise.
However, I will be applying a method of feature_selection to reduce our total features. 

The attribute we want to use is called 'Count' or 'cnt' (as per the dataset) 
below we will doing:

* create a train test split and remove the 'cnt' attribute from the X, y variables
* run a model and apply feature selection based on the model
* run another model


```python
y, X = dataset['cnt'], dataset.drop(['cnt'], axis= 1) # X will be the class var
print(y.head())
y = y[:, None] # this step allows us to even the number of labels and samples
y.shape
```

```python
# we should also drop the 'dteday' column as it serves no purpose in the later stages.
X = X.drop(['dteday'], axis = 1)
X = X.drop(['instant'], axis = 1)
print(X.columns)
X.shape
```

```python
clf = LinearSVC()
clf.fit(X, y)

fs_model = SelectFromModel(clf, prefit=True)
X = fs_model.transform(X)

X.shape
```

```python
# X is now reduced from 14 attributes to 6. 
# Given the many different ways of feature selection, this
# may or may not be a right way for this type of data.
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 420)
# The reason for using a 80/20 train/test split is because we have 17,000+ datapoints, which gives us enough data to 
# somewhat accurately train the model. 
```

```python
# Linear Regression algorithm works best on data with a linear distribution. 
# Given the explopratory data analysis we did, this may not be the best 
# algorith to apply.
model = LinearRegression()

model.fit(train_X, train_y)

# store the predictions
pred = model.predict(test_X)
```

```python
# plc == placeholder
plc = list(zip(pred, test_y))

#error list to see the difference between the actual value and predicted value
# more detailed analysis when loaded in to pandas
error = [i -j for i, j in plc]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 

print(mean_absolute_error(test_y, pred))

# calculate the ROOT Mean Square Error
# the MSE is a good metric by RMSE provides a more closer attachment
print(mean_squared_error(test_y, pred)**0.5)
```

```python
error_df = pd.DataFrame(error)
print(error_df.median())
error_df.describe()

```

```python
plt.plot(error, ',')
```

Using the results from our feature selection and model (linear Regression) we can see that most of our prediction are clumped quite close to the 0 differernce (from the graph).
When using proper metrics such as RMSE and MAE we can precieve the impact of Linear Regression on our Dataset.
